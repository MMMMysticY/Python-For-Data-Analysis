```python
import pandas as pd
import numpy as np
import temp as temp
from pandas import Series,DataFrame
```

```python
close_px_all = pd.read_csv('../examples/stock_px.csv',
                           parse_dates=True,index_col=0)
close_px = close_px_all[['AAPL','MSFT','XOM']]
close_px
```

```python
close_px = close_px.resample('B').ffill()
close_px
```

```python
close_px['AAPL'].plot()
```

```python
close_px.loc['2009'].plot()
#选取2009年的 这个就很有意思 timestamp可以直接选年的
```

```python
close_px['AAPL'].loc['01-2011':'03-2011'].plot()
#更有意思了 timestamp可以用区间选
```

```python
apple_q = close_px['AAPL'].resample('Q-DEC').ffill()
apple_q
```

```python
apple_q.loc['2009':].plot()
```
