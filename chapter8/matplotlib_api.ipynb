```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
```

```python
import matplotlib.pyplot as plt
import matplotlib
from numpy.random import randn
# matplotlib.use('TKagg')
```

```python
#绘制子图 subplot类方法


#方法一：（plt的subplot方法）
#此方法是将plt对象看做整体 没有明确实例化子图对象

fig = plt.figure()
#matplotlib的图像都位于Figure对象中
x1 = randn(50).cumsum()
plt.subplot(221)
#221意味所有图形共分2行2列 4个子图 此子图为第一个
plt.plot(x1,'k--')

x2 = randn(100)
plt.subplot(222)
plt.hist(randn(50),bins=20,color='k')

plt.subplot(223)
plt.scatter(np.arange(30),np.arange(30)+3*randn(30))

plt.show()
```

```python
#方法二：add_subplot方法 和书上类似
#此方法实例化的不同的子图对象，让子图对象调用画图方法
fig2 = plt.figure()
ax1 = fig2.add_subplot(221)
ax1.plot(x1,'k-')

ax2 = fig2.add_subplot(222)
ax2.hist(randn(50),bins=20,color='k')

ax3 = fig2.add_subplot(223)
ax3.scatter(np.arange(30),np.arange(30)+3*randn(30))

plt.show()
```

```python
#方法三：plt.subplots方法 书上第二种方法
#此方法将多个子图对象变为一个整体，再让它们各自调用画图方法
fig3, axes = plt.subplots(2,2)
axes[0][0].plot(x1,'k-')

axes[0][1].hist(randn(50),bins=20,color='k')

axes[1][0].scatter(np.arange(30),np.arange(30)+3*randn(30))

plt.show()
```

```python
#修改子图间的间距
fig4,axes = plt.subplots(2,2,sharex='all',sharey='all')
for i in range(2):
    for j in range(2):
        axes[i][j].hist(randn(500),bins=50,color='k')
plt.subplots_adjust(wspace=0,hspace=0)
#subplots_adjust可以改变子图间的关系 wspace和hspace用于控制宽度和高度的百分比
plt.show()
```

```python
#颜色 标记 线型
plt.plot(randn(30).cumsum(),color='k',linestyle='dashed',marker='o')
plt.show()
#color参数确定颜色， linestyle确定线条风格 marker可以加上标记
```

```python
data = randn(30).cumsum()
plt.plot(data,'k--')
plt.show()
```

```python
plt.plot(data,'k-',drawstyle='steps-post')
plt.show()
#默认是采用线性插值方法进行画图 使用drawstyle参数可以进行修改
```

```python
#刻度 标签 图例
plt.xlim()
#使用xlim()不带参数，则是一种get方法，获取当前使用的x轴刻度范围
```

```python
plt.xlim([-10,10])
plt.plot(data,'k-',drawstyle='steps-post')
plt.show()
#使用xlim带区间参数，就是一种set方法，设置画图的x轴刻度范围
#注意对于subplot来说 每个子图都有get_xlim和set_xlim方法 用于修改各自的x轴范围(ylim 同理)
```

```python
fig5 = plt.figure()
ax5 = fig5.add_subplot(111)
ax5.plot(randn(1000).cumsum())
```

```python
fig5 = plt.figure()
ax5 = fig5.add_subplot(111)
ticks = ax5.set_xticks([0,250,500,750,1000])
ax5.plot(randn(1000).cumsum())
#使用set_xticks可以设置x轴的刻度显示位置
```

```python
fig5 = plt.figure()
ax5 = fig5.add_subplot(111)
ticks = ax5.set_xticks([0,250,500,750,1000])
labels = ax5.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize='small')
ax5.set_title('not my first plot')
ax5.set_xlabel('Stage')
ax5.plot(randn(1000).cumsum())
#同时可以使用set_xticklabels设置名称 旋角 字体等
#使用set_title设置名称 set_xlabel设置x轴的名称等
```

```python
fig6 = plt.figure()
ax6 = fig6.add_subplot(111)
ax6.plot(randn(1000).cumsum(),'k',label='one')
ax6.plot(randn(1000).cumsum(),'k--',label='two')
ax6.plot(randn(1000).cumsum(),'k.',label='three')
ax6.legend(loc='best')
#使用label可以给同一个图的每个曲线进行命名 最后使用legend显示 loc=best会找到最佳位置
```

```python
#text可以将文本绘制在制定坐标(x,y)

fig6 = plt.figure()
ax6 = fig6.add_subplot(111)
ax6.plot(randn(1000).cumsum(),'k',label='one')
ax6.plot(randn(1000).cumsum(),'k--',label='two')
ax6.plot(randn(1000).cumsum(),'k.',label='three')
ax6.text(100,0,'Hello',fontsize=10)
ax6.legend(loc='best')
```

```python
#画数学图形
fig7 = plt.figure()
ax7 = fig7.add_subplot(111)
rect = plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.5)
circ = plt.Circle((0.7,0.2),0.15,color='b',alpha=0.5)
pgon = plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='g',alpha=0.5)
ax7.add_patch(rect)
ax7.add_patch(circ)
ax7.add_patch(pgon)
#plt中有几个能画图形的 如Rectangle circle polygon等
#但是完整的在matplotlib.patches中
```
