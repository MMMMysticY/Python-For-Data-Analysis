```python
fec = fec[fec['contb_receipt_amt']>0]
fec
```

```python
fec_mrbo = fec[fec['cand_nm'].isin(['Obama, Barack', 'Romney, Mitt'])]
fec_mrbo
#这个是挑出了两个主要候选人的信息 isin值的注意
```

```python
fec['contbr_occupation'].value_counts()
#如果按照职业进行分析的话 第二大的部分 information requested是未知的
#再清洗一下
```

```python
occ_mapping = {
    'INFORMATION REQUESTED':'NOT PROVIDED',
    'INFORMATION REQUESTED PER BEST EFFORTS':'NOT PROVIDED',
    'INFORMATION REQUESTED (BEST EFFORTS)':'NOT PROVIDED'
}
```

```python
f = lambda x:occ_mapping.get(x,x)
fec['contbr_occupation'] = fec['contbr_occupation'].map(f)
fec
```

```python
emp_mapping = {
    'INFORMATION REQUESTED':'NOT PROVIDED',
    'INFORMATION REQUESTED PER BEST EFFORTS':'NOT PROVIDED',
    'SELF':'SELF-EMPLOYED',
    'SELF EMPLOYED':'SELF-EMPLOYED'
}
f1 = lambda x:emp_mapping.get(x,x)
fec['contbr_employer'] = fec['contbr_employer'].map(f1)
fec

#总结一下 这个使用字典的get方法进行map 然后第二个参数是如果没有映射关系就保留 很巧妙
```

```python
by_occupation = fec.pivot_table('contb_receipt_amt',index='contbr_occupation',
                                columns='party',aggfunc='sum')
by_occupation
#按照职业和党派做透视图
```

```python
over_2mm = by_occupation[by_occupation.sum(1)>200000]
over_2mm
#按照职业进行相加 选出出资额高于200万美元的部分
```

```python
over_2mm.plot(kind='barh')
```

```python
def get_top_amounts(group,key,n=3):
    totals = group.groupby(key)['contb_receipt_amt'].sum()
    # print("-------------------")
    # print(type(totals))
    # print(totals.sort_values(ascending=False))
    # print('----------------------')
    # print(totals.sort_values(ascending=False)[:n])
    # print('#######################')
    return totals.sort_values(ascending=False)[:n]
```

```python
grouped = fec_mrbo.groupby('cand_nm')
grouped.size()
```

```python
grouped.apply(get_top_amounts,'contbr_occupation',n=5)
#可以看到都是上流职业 牛逼
```

```python
grouped.apply(get_top_amounts,'contbr_employer',n=5)
#美帝还是牛逼
```

```python
bins = np.array([0,1,10,100,1000,10000,100000,1000000,10000000])
labels = pd.cut(fec_mrbo['contb_receipt_amt'],bins=bins)
labels
```

```python
grouped1 = fec_mrbo.groupby(['cand_nm',labels])
grouped1.size()
```

```python
grouped1.size().unstack()
#奥巴马的支持者是有钱
```

```python
bucket_sum = grouped1['contb_receipt_amt'].sum().unstack(0)
bucket_sum
#把不同分组的加起来
```

```python
grouped1['contb_receipt_amt'].sum()
```

```python
normed_sums = bucket_sum.div(bucket_sum.sum(axis=1),axis=0)
normed_sums
#按照不同的收入群体所占的各自的比例
```

```python
normed_sums[:-2].plot(kind='barh',stacked=True)
#奥巴马还是比较厉害 基本是全线领先 除了中产阶级？
```
