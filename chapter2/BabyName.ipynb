```python
years = range(1880, 2011)
pieces = []
columns = ['name', 'sex', 'births']
root_path = '../data/babynames/'
for year in years:
    path = root_path+'yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    frame['year'] = year
    pieces.append(frame)
names = pd.concat(pieces, ignore_index = True) #concat默认是按行将多个DataFrame组合到一起的 设置ignore_index为了去除read_csv的原始行号
names
```

```python
def add_prop(group):
    group['prop'] = group.births/group.births.sum()
    return group

names = names.groupby(['year','sex']).apply(add_prop)
names
```

```python
import numpy as np
np.allclose(names.groupby(['year','sex']).prop.sum(), 1) #检查prop的和是否为1
```

```python
def get_top1000(group):
    return group.sort_values('births', ascending=False)[:1000]

grouped = names.groupby(['year','sex'])
top1000 = grouped.apply(get_top1000)
top1000
```

```python
boys = top1000[top1000['sex'] == 'M']
girls = top1000[top1000['sex'] == 'F']
```

```python
top1000.index = top1000.index.droplevel()
top1000
```

```python
total_births = top1000.pivot_table('births', index = 'year', columns = 'name', aggfunc = sum)
total_births
```
