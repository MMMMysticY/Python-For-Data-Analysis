```python
import pandas as pd
root_path = '../data/movielens/'
unames = ['userId', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table(root_path+'users.dat', sep = '::', header=None, names=unames)
users
```

```python
rname = ['userId', 'movieId', 'rating', 'timestamp']
ratings = pd.read_table(root_path+'ratings.dat',sep = '::', header=None, names=rname)
ratings
```

```python
mname = ['movieId', 'title', 'genres']
movies = pd.read_table(root_path+'movies.dat',sep = '::', header=None, names=mname)
movies
```

```python
data = pd.merge(pd.merge(ratings, users), movies)  #有相同的列merge会自动合并
```

```python
data
```

```python
mean_rating = data.pivot_table('rating', index = 'title', columns = 'gender', aggfunc = 'mean')
```

```python
mean_rating
```

```python
rating_by_title = data.groupby('title').size()
rating_by_title
```

```python
print(type(rating_by_title))
```

```python
print(rating_by_title.index)
```

```python
active_title = rating_by_title.index[rating_by_title>250]
active_title
```

```python
mean_rating = mean_rating.loc[active_title]
mean_rating
```

```python
top_female_rating = mean_rating.sort_values('F', ascending = False)
top_female_rating[:10]
```

```python
mean_rating['diff'] = mean_rating['M']-mean_rating['F']
mean_rating
```

```python
sorted_by_diff = mean_rating.sort_values('diff')
sorted_by_diff
```
